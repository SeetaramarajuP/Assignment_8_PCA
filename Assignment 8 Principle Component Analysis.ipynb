```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv('wine.csv')
wine.head()
```

```python
data=wine.iloc[:,1:]
data
```

```python
values=data.values
values
```

```python
data_norm=scale(values)
data_norm
```

```python
pca=PCA(n_components=13)
pca_values=pca.fit_transform(data_norm)
pca_values
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals=3)*100)
var1
```

```python
plt.plot(var1,color='red')
```

```python
pca_values[:,:3]
```

```python
f_wine=pd.DataFrame(pca_values[:,:3],columns=['pc1','pc2','pc3'])
f_wine
```

K Mean

```python
from sklearn.cluster import KMeans
```

```python
WCSS=[]
for i in range (1,5):
    clf=KMeans(n_clusters=i)
    clf.fit(f_wine)
    WCSS.append(clf.inertia_)
plt.plot(range(1,5),WCSS)
plt.show()
```

```python
df=KMeans(n_clusters=3)
y_KMeans=clf.fit_predict(f_wine)
y_KMeans
```

```python
md=pd.Series(y_KMeans)
wine['Kmean']=md
wine
```

```python
f_wine.plot(x='pc1',y='pc2',c=clf.labels_,kind='scatter',s=20,cmap=plt.cm.coolwarm)
```

H-Cluster

```python
def norm_wine(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
wine_norm=norm_wine(wine.iloc[:,1:])
wine_norm
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(f_wine,method='average',metric='euclidean')
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=3,linkage='average',affinity='euclidean').fit(f_wine)
```

```python
cluster_labels=pd.Series(h_complete.labels_)
wine['H-Cluster']=cluster_labels
wine
```
